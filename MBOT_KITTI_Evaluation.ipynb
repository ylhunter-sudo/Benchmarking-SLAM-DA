```python
#@title 3) Trajectory and GT error plots

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


def load_kitti_gt_xyz(pose_file: Path):
    arr = np.loadtxt(pose_file)
    if arr.ndim == 1:
        arr = arr[None, :]
    mats = arr.reshape(-1, 3, 4)
    return mats[:, :, 3]


def load_times(times_file: Path):
    return np.loadtxt(times_file)


def load_traj_xyz_auto(traj_file: Path):
    arr = np.loadtxt(traj_file)
    if arr.ndim == 1:
        arr = arr[None, :]

    # TUM / EuRoC-like: timestamp tx ty tz ...
    if arr.shape[1] >= 4:
        t = arr[:, 0].astype(float)
        xyz = arr[:, 1:4].astype(float)
        if np.nanmax(t) > 1e6:
            t = t / 1e9
        return t, xyz

    raise ValueError(f"Unsupported trajectory shape {arr.shape} for {traj_file}")


def match_to_kitti_frames(pred_t, pred_xyz, kitti_times, gt_xyz):
    idx = np.searchsorted(kitti_times, pred_t)
    idx = np.clip(idx, 0, len(kitti_times) - 1)

    prev_idx = np.clip(idx - 1, 0, len(kitti_times) - 1)
    use_prev = np.abs(kitti_times[prev_idx] - pred_t) < np.abs(kitti_times[idx] - pred_t)
    idx[use_prev] = prev_idx[use_prev]

    seen = {}
    for i, j in enumerate(idx):
        err = abs(kitti_times[j] - pred_t[i])
        if j not in seen or err < seen[j][0]:
            seen[j] = (err, i)

    keep_pred_idx = np.array(sorted(v[1] for v in seen.values()), dtype=int)
    keep_gt_idx = idx[keep_pred_idx]

    pred_xyz_matched = pred_xyz[keep_pred_idx]
    gt_xyz_matched = gt_xyz[keep_gt_idx]
    return pred_xyz_matched, gt_xyz_matched, keep_gt_idx


def umeyama_alignment(X, Y, with_scale=True):
    assert X.shape == Y.shape
    n = X.shape[0]

    mean_X = X.mean(axis=0)
    mean_Y = Y.mean(axis=0)

    Xc = X - mean_X
    Yc = Y - mean_Y

    cov = (Yc.T @ Xc) / n
    U, D, Vt = np.linalg.svd(cov)

    S = np.eye(3)
    if np.linalg.det(U) * np.linalg.det(Vt) < 0:
        S[-1, -1] = -1

    R = U @ S @ Vt

    if with_scale:
        var_X = np.mean(np.sum(Xc**2, axis=1))
        s = np.trace(np.diag(D) @ S) / var_X
    else:
        s = 1.0

    t = mean_Y - s * (R @ mean_X)
    return s, R, t


def apply_sim3(X, s, R, t):
    return (s * (R @ X.T)).T + t


def evaluate_one_run(traj_file: Path, gt_xyz, kitti_times):
    pred_t, pred_xyz = load_traj_xyz_auto(traj_file)
    pred_xyz, gt_match, gt_idx = match_to_kitti_frames(pred_t, pred_xyz, kitti_times, gt_xyz)

    if len(pred_xyz) < 5:
        raise RuntimeError(f"Too few matched points: {traj_file}")

    s, R, t = umeyama_alignment(pred_xyz, gt_match, with_scale=True)
    pred_aligned = apply_sim3(pred_xyz, s, R, t)
    err = np.linalg.norm(pred_aligned - gt_match, axis=1)

    return {
        "pred_aligned": pred_aligned,
        "gt_match": gt_match,
        "gt_idx": gt_idx,
        "err": err,
        "rmse": float(np.sqrt(np.mean(err**2))),
        "mean": float(np.mean(err)),
        "median": float(np.median(err)),
        "max": float(np.max(err)),
        "scale": float(s),
        "n": int(len(pred_aligned)),
    }


seq = DISPLAY_SEQUENCE
gt_pose_file = GT_POSE_FILES[seq]
times_file = TIMES_FILES[seq]

assert gt_pose_file.exists(), f"Missing GT file: {gt_pose_file}"
assert times_file.exists(), f"Missing times file: {times_file}"

gt_xyz = load_kitti_gt_xyz(gt_pose_file)
times = load_times(times_file)

results = {}

for run_name in DISPLAY_RUNS:
    traj_file = RUNS[run_name][SEQUENCES.index(seq)]
    print(f"{run_name}: {traj_file} | exists={traj_file.exists()}")
    if traj_file.exists():
        try:
            results[run_name] = evaluate_one_run(traj_file, gt_xyz, times)
        except Exception as e:
            print(f"Skipping {run_name} due to error: {e}")

assert results, "No valid trajectories loaded for display"

if SHOW_XZ or SHOW_XY:
    ncols = int(SHOW_XZ) + int(SHOW_XY)
    plt.figure(figsize=(6 * ncols, 5))

    plot_i = 1

    if SHOW_XZ:
        plt.subplot(1, ncols, plot_i)
        if SHOW_GT:
            gt_style = get_plot_style("GT")
            plt.plot(gt_xyz[:, 0], gt_xyz[:, 2], label='GT', **gt_style)

            gt_color = gt_style.get("color", "k")
            plt.scatter(gt_xyz[0, 0], gt_xyz[0, 2], color=gt_color, marker='o', s=70, zorder=5)
            plt.scatter(gt_xyz[-1, 0], gt_xyz[-1, 2], color=gt_color, marker='X', s=90, zorder=6)

        for run_name, res in results.items():
            run_style = get_plot_style(run_name)
            traj = res["pred_aligned"]

            plt.plot(
                traj[:, 0],
                traj[:, 2],
                label=RUN_LABELS.get(run_name, run_name),
                **run_style
            )

            run_color = run_style.get("color", None)
            plt.scatter(traj[0, 0], traj[0, 2], color=run_color, marker='o', s=55, zorder=5)
            plt.scatter(traj[-1, 0], traj[-1, 2], color=run_color, marker='X', s=75, zorder=6)

        plt.xlabel("x [m]")
        plt.ylabel("z [m]")
        plt.title(f"Sequence {seq}: XZ trajectory")
        plt.axis("equal")
        plt.grid(True)
        plt.legend()
        plot_i += 1

    if SHOW_XY:
        plt.subplot(1, ncols, plot_i)
        if SHOW_GT:
            gt_style = get_plot_style("GT")
            plt.plot(gt_xyz[:, 0], gt_xyz[:, 1], label='GT', **gt_style)

            gt_color = gt_style.get("color", "k")
            plt.scatter(gt_xyz[0, 0], gt_xyz[0, 1], color=gt_color, marker='o', s=70, zorder=5)
            plt.scatter(gt_xyz[-1, 0], gt_xyz[-1, 1], color=gt_color, marker='X', s=90, zorder=6)

        for run_name, res in results.items():
            run_style = get_plot_style(run_name)
            traj = res["pred_aligned"]

            plt.plot(
                traj[:, 0],
                traj[:, 1],
                label=RUN_LABELS.get(run_name, run_name),
                **run_style
            )

            run_color = run_style.get("color", None)
            plt.scatter(traj[0, 0], traj[0, 1], color=run_color, marker='o', s=55, zorder=5)
            plt.scatter(traj[-1, 0], traj[-1, 1], color=run_color, marker='X', s=75, zorder=6)

        plt.xlabel("x [m]")
        plt.ylabel("y [m]")
        plt.title(f"Sequence {seq}: XY trajectory")
        plt.axis("equal")
        plt.grid(True)
        plt.legend()

    plt.tight_layout()
    plt.show()

if SHOW_ERROR:
    plt.figure(figsize=(10, 4))
    for run_name, res in results.items():
        run_style = get_plot_style(run_name)
        plt.plot(res["gt_idx"], res["err"], label=RUN_LABELS.get(run_name, run_name), **run_style)
    plt.xlabel("KITTI frame index")
    plt.ylabel("Position error [m]")
    plt.title(f"Sequence {seq}: aligned position error")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

summary_rows = []
for run_name, res in results.items():
    summary_rows.append({
        "run": run_name,
        "label": RUN_LABELS.get(run_name, run_name),
        "matched_points": res["n"],
        "rmse_m": res["rmse"],
        "mean_m": res["mean"],
        "median_m": res["median"],
        "max_m": res["max"],
        "scale": res["scale"],
    })

import pandas as pd
display(pd.DataFrame(summary_rows).sort_values("rmse_m"))
```

```python
#@title 4) Comprehensive Summary, Report, and Charts

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Helper functions for calculations
def compute_path_length(xyz):
    if len(xyz) < 2:
        return 0.0
    return float(np.sum(np.linalg.norm(np.diff(xyz, axis=0), axis=1)))

def compute_displacement(xyz):
    if len(xyz) < 2:
        return 0.0
    return float(np.linalg.norm(xyz[-1] - xyz[0]))

def load_gt_for_seq(seq):
    gt_pose_file = GT_POSE_FILES[seq]
    times_file = TIMES_FILES[seq]
    gt_xyz = load_kitti_gt_xyz(gt_pose_file)
    gt_times = load_times(times_file)
    return gt_xyz, gt_times

def evaluate_report_row(seq, run_name, traj_file):
    if not traj_file.exists():
        return None

    try:
        gt_xyz, gt_times = load_gt_for_seq(seq)

        pred_t, pred_xyz = load_traj_xyz_auto(traj_file)
        pred_xyz_m, gt_xyz_m, gt_idx = match_to_kitti_frames(pred_t, pred_xyz, gt_times, gt_xyz)

        if len(pred_xyz_m) < 5:
            return None

        pred_path = compute_path_length(pred_xyz_m)
        pred_disp = compute_displacement(pred_xyz_m)

        gt_path = compute_path_length(gt_xyz_m)
        gt_disp = compute_displacement(gt_xyz_m)

        duration_s = float(pred_t.max() - pred_t.min()) if len(pred_t) > 1 else 0.0

        s, R, t = umeyama_alignment(pred_xyz_m, gt_xyz_m, with_scale=True)
        pred_aligned = apply_sim3(pred_xyz_m, s, R, t)
        err = np.linalg.norm(pred_aligned - gt_xyz_m, axis=1)

        return {
            "sequence": seq,
            "run": run_name,
            "label": RUN_LABELS.get(run_name, run_name),
            "poses": int(len(pred_xyz)),
            "matched_pts": int(len(pred_xyz_m)),
            "duration_s": round(duration_s, 2),
            "path_ratio_vs_gt": round(pred_path / gt_path, 3) if gt_path > 0 else np.nan,
            "disp_ratio_vs_gt": round(pred_disp / gt_disp, 3) if gt_disp > 0 else np.nan,
            "rmse_m": round(float(np.sqrt(np.mean(err**2))), 3),
            "median_m": round(float(np.median(err)), 3),
            "max_m": round(float(np.max(err)), 3),
            "sim3_scale": round(float(s), 3),
        }
    except Exception as e:
        print(f"Error evaluating {run_name} for sequence {seq}: {e}")
        return None

# --- Data Generation for the detailed report table ---
report_rows = []

for seq in SEQUENCES:
    for run_name, traj_list in RUNS.items():
        traj_file_idx = SEQUENCES.index(seq)
        traj_file = Path(traj_list[traj_file_idx])
        row = evaluate_report_row(seq, run_name, traj_file)
        if row is not None:
            report_rows.append(row)

report_df = pd.DataFrame(report_rows)

# Ordering and display for the detailed report
report_df = report_df.sort_values(["sequence", "label"]).reset_index(drop=True)
print("Detailed Report Table (per-sequence trajectory summary):")
display(report_df)

# Optional export for report writing
out_csv_report = DRIVE_ROOT / "trajectory_report_table.csv"
report_df.to_csv(out_csv_report, index=False)
print(f"Saved detailed report to: {out_csv_report}")


# --- Plotting across all sequences and run types ---


fig = plt.figure(figsize=(25, 15))

# 1) RMSE by sequence
ax1 = plt.subplot(3, 2, 1)
for run_name in RUNS.keys():
    sub = report_df[report_df["run"] == run_name]
    style = get_plot_style(run_name)
    ax1.plot(sub["sequence"], sub["rmse_m"], marker='o', label=RUN_LABELS.get(run_name, run_name),
             color=style["color"], linestyle=style["linestyle"])
ax1.set_title("RMSE by sequence")
ax1.set_xlabel("Sequence")
ax1.set_ylabel("RMSE [m]")
ax1.grid(True)
ax1.legend()

# 2) Median error by sequence
ax2 = plt.subplot(3, 2, 2)
for run_name in RUNS.keys():
    sub = report_df[report_df["run"] == run_name]
    style = get_plot_style(run_name)
    ax2.plot(sub["sequence"], sub["median_m"], marker='o', label=RUN_LABELS.get(run_name, run_name),
             color=style["color"], linestyle=style["linestyle"])
ax2.set_title("Median error by sequence")
ax2.set_xlabel("Sequence")
ax2.set_ylabel("Median error [m]")
ax2.grid(True)
ax2.legend()

# 3) Matched points by sequence
ax3 = plt.subplot(3, 2, 3)
for run_name in RUNS.keys():
    sub = report_df[report_df["run"] == run_name]
    style = get_plot_style(run_name)
    ax3.plot(sub["sequence"], sub["matched_pts"], marker='o', label=RUN_LABELS.get(run_name, run_name),
             color=style["color"], linestyle=style["linestyle"])
ax3.set_title("Matched trajectory points by sequence")
ax3.set_xlabel("Sequence")
ax3.set_ylabel("Matched points")
ax3.grid(True)
ax3.legend()

# 4) Path Ratio vs GT by sequence
ax4 = plt.subplot(3, 2, 4)
for run_name in RUNS.keys():
    sub = report_df[report_df["run"] == run_name]
    style = get_plot_style(run_name)
    ax4.plot(sub["sequence"], sub["path_ratio_vs_gt"], marker='o', label=RUN_LABELS.get(run_name, run_name),
             color=style["color"], linestyle=style["linestyle"])
ax4.set_title("Path Ratio vs GT by sequence")
ax4.set_xlabel("Sequence")
ax4.set_ylabel("Path Ratio (Pred / GT)")
ax4.axhline(y=1.0, color='gray', linestyle='--', linewidth=1, label='Ideal (1.0)')
ax4.grid(True)
ax4.legend()

# 5) Sim3 Scale by sequence
ax5 = plt.subplot(3, 2, 5)
for run_name in RUNS.keys():
    sub = report_df[report_df["run"] == run_name]
    style = get_plot_style(run_name)
    ax5.plot(sub["sequence"], sub["sim3_scale"], marker='o', label=RUN_LABELS.get(run_name, run_name),
             color=style["color"], linestyle=style["linestyle"])
ax5.set_title("Sim3 Scale by sequence")
ax5.set_xlabel("Sequence")
ax5.set_ylabel("Sim3 Scale")
ax5.axhline(y=1.0, color='gray', linestyle='--', linewidth=1, label='Ideal (1.0)')
ax5.grid(True)
ax5.legend()

plt.tight_layout()
plt.show()


print("\nAverage Metrics by Run:")
avg_report_df = (
    report_df
    .groupby(["run", "label"], as_index=False)
    .agg({
        "rmse_m": "mean",
        "median_m": "mean",
        "max_m": "mean",
        "path_ratio_vs_gt": "mean",
        "disp_ratio_vs_gt": "mean",
        "sim3_scale": "mean",
        "matched_pts": "mean",
        "poses": "mean",
    })
    .sort_values("rmse_m")
)

# round for presentation
for col in ["rmse_m",
            "median_m", "max_m",
            "path_ratio_vs_gt", "disp_ratio_vs_gt", "sim3_scale",
            "matched_pts", "poses"]:
    avg_report_df[col] = avg_report_df[col].astype(float).round(3)

display(avg_report_df)

out_csv_avg = DRIVE_ROOT / "trajectory_report_avg_by_run.csv"
avg_report_df.to_csv(out_csv_avg, index=False)
print(f"Saved average report to: {out_csv_avg}")


plt.figure(figsize=(4, 4))
bar_colors = [get_plot_style(run)["color"] for run in avg_report_df["run"]]
plt.bar(avg_report_df["label"], avg_report_df["rmse_m"], color=bar_colors)
plt.title("Average RMSE across sequences")
plt.ylabel("Average RMSE [m]")
plt.tick_params(axis='x', rotation=25)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
